````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas import Series, DataFrame
import seaborn as sns
from IPython.display import display
import time
from scipy import stats
import hashlib
from datetime import datetime
````

````python
def get_single_file(filename):
    df1 = pd.read_csv(filename, sep=' ')
    df1 = df1.set_axis(["ATime","RTime", "Previous", "Current", "Signal", "Seq", "Source"], axis=1)
    freq_map = {2412: 1, 2437: 6, 2462: 11}
    df1 = df1.replace({"Previous": freq_map})
    df1 = df1.replace({"Current": freq_map})
    df1["Tag"] = df1["Previous"].astype(str) + "->" + df1["Current"].astype(str)
    df1["ITime"] = df1["ATime"].diff()
    df1 = df1.astype({"ATime": np.float32, "RTime": np.float32, "Signal": np.float32, "ITime": np.float32})
    return df1.loc[1:, :]



# each of the parameters should be a list of values
# might be a more efficient way with os.walk?
def get_data(days, months, years):
    start = time.time()
    final_res = []
    for year in years:
        for month in months:
            for day in days:
                single_day = []
                for sample in range(24):
                    try:
                        df = get_single_file(f"../data/text/hourly/{day:02}_{month}_{year}/sample{sample:02}.txt")
                        df["Sample"] = sample
                        single_day.append(df)
                        
                    except Exception as e:
                        print(f"Error with {day}/{month}/{year}: {e}")
                        continue
                    
                    
                x = pd.concat(single_day)
                
                
                x["Date"] = datetime.strptime(f"{day}/{month}/{year}", '%d/%m/%y').date()
                x["CType"] = "Different" 
                x.loc[x["Previous"] == x["Current"], "CType"] = "Same"
                
                x = x[["ATime", "RTime", "ITime", "Date", "Previous", "Current", "Tag", "CType", "Signal", "Seq", "Source"]]
                final_res.append(x)
                
    end = time.time()
    print("Total Time: ", end - start)
    return pd.concat(final_res)

df = get_data(range(1,5,1), [12], [22])
````

# Identifying Devices
Now that we have some idea that phones should have unique distributions (we'll need more phones to make sure), we can get some ideas going for testing whether they are the same device. There are several things to consider for our solution:

1. Will we be doing batch processing or online processing?
2. Do we only care about determining the number of devices, or also tracking particular devices?
3. Will we only use timing? Or will we also use other information found in the probes?

Assuming we can use the sequence number as well, we could build a rather simple decision tree model that uses the MAC address, the Sequence number, and the Interarrival time.

In code, this would simply be a check on the MAC address followed by clustering using time of the probe request and the sequence number in the management frame. We must also acknowledge that sequence numbers are an uneccessary security flaw and will likely be randomized in the future. This leaves only the timing behind for us to use. 

The main idea is to seperate the single long chain of packets into seperate chains where each chain represents a single device probing over time. The information we would use is:
1. The arrival time of the packet
2. The channel of the packet
3. The length of each existing chain
4. The MAC address
5. The arrival time of the LAST packet in each chain
6. The channel of the LAST packet in each chain


The only piece of information we need is the number of packets sent per scan. This value likely differs across devices, but it is still critical. Without this piece of information, we would always assume a packet is related with a recent packet, despite it potentially being part of a chain that just started a new scan. We need to find a way to balance chain length, arrival time, and channel used. The chain length parameter should lower the probability of a packet being attached to a chain that is near that average length (or potentially even minimum), but should reset once we hit that minimum and have spent 160 seconds since the arrival of the last packet in a chain.

````python
# Finding the number of packets sent out per scan per device
def packets_per_scan(group):
    x = group["RTime"].to_numpy()
    diff = np.diff(x)
    seperators = np.where(diff > 5)[0]
    scan_sizes = np.abs(np.diff(seperators))
    name = pd.unique(group.loc[:, "Source"])[0]
    sns.histplot(scan_sizes, bins=np.arange(0,70,2)).set(title=name, xlabel="Packets Per Scan")
    plt.show()
    
    y = DataFrame(scan_sizes)
    # display(y.describe())
    
    
    
    
df.groupby("Source").apply(packets_per_scan)
````

The number of packets per scan seems to be bell shaped for each device, however the means vary enough that we cannot use this information reliably until we have a reasonably sized chain that we know is a single device.

We also know that each scan takes only around a second. Thus the probability of assigning a packet to a chain where the time since the last packet is between 1 and 20 seconds should be very low. Another assumption we might make is that each phone is in the last state of the scan plan since it is unlikely that people are turning their wifi off and on.

# Features

## Arrival Time

We have several windows of interest, but we'll start with the simplest:

```
X % 160 < 1.5 = [1,0]
1.5 > X % 160 > 160 = [0,1]
```

The probability of `[1,0]` should be 1 and the probability of `[0,1]` should be 0.

## Channel

We know that phones typically stay on the same channel, but we haven't really examined it in detail. Let's take a look now.




<!-- Different = 0

````python
def packets_per_channel(group):
    seperators = np.sort(group[group.Previous != group.Current].index.to_numpy()) + 1
    
    packets = np.diff(seperators)
    packets = packets[np.where(packets < 80)]
    
    name = pd.unique(group.loc[:, "Source"])[0]
    sns.histplot(packets, bins=np.arange(0,30,1)).set(title=name, xlabel='Consecutive Packets on the same Channel')
    plt.show()
    
    g = sns.displot(packets, kind="ecdf")
    plt.show()
    
    
    y = DataFrame(packets)
    # display(y.describe())
    
df.groupby("Source").apply(packets_per_channel)
````

## MAC Address

OK, this isn't timing related, but if the MAC isn't randomized, let's not even use the model.

## Features for non ML Model
1) MAC Address
2) Arrival Time
3) Channel Packet Sent On


Pretty much, we'll just have a basic decision tree that decides if the current packet should be placed on an existing chain or if it should start a new one.

````python
from enum import Enum


class Channel(Enum):
    ONE = 1
    SIX = 6
    ELEVEN = 11
    
    
    
    
class Packet():
    time = None
    channel = None
    address = None
    
    def __init__(self, time: float, channel: int, address: str):
        self.time = time
        self.channel = Channel(channel)
        self.address = address
    


class Chain():
    
    address = None # address is only used in the case where MAC randomiation is not used
    length = None
    time_of_last_packet = None
    channel_of_last_packet = None
    packets = []
    
    
    def __init__(self, mac: str):
        if mac is not None and not self.check_mac(mac):
            self.address = mac
        
        
    def check_mac(self, string):
        try:
            x = bytearray(string, encoding="ascii")[1]
        except:
            return True
    
        y = (x >> 1) & 1
        return y == 1
    
    
    def add_packet_to_chain(self, packet: Packet):
        self.channel_of_last_packet = packet.channel
        self.time_of_last_packet = packet.time
        self.packets.append(packet)
        
        
class ChainList():
    
    
    def __init__(self):
        pass
    
    def run(self, file):
        f = open(file)
        lines = [x.strip() for x in f]
        mapping = {'2412': 1, '2437': 6, '2462': 11}
        
        for line in lines:
            
            atime, rtime, start, end, signal, seq, source = line.split(" ")
            atime, rtime, signal, seq = float(atime), float(rtime), int(signal), int(seq)
            start, end = mapping[start], mapping[end]
            
            packet: Packet = Packet(rtime, end, source)
            chain: Chain = self.find_chain(packet)
            chain.add_packet_to_chain(packet)
            
            
            
    def find_chain(self, packet: Packet):
        
        if len(self.chains) == 0:
            chain: Chain = Chain(packet.address)
            self.chains.append(chain)
            return chain
        
        
        # first we just check to see if the address has been identified
        for x in self.chains:
            if x.address == packet.address:
                return x
            
    
        # next, we'll use the channel 
        
        
        
        # finally we'll use the interarrival time
        
        
        # if we are still unsure, we'll create a new one
        chain: Chain = Chain(packet.address)
        self.chains.append(chain)
        return chain
        
        
````

# Some Random Stuff

While reading through the Apple WiFi security [page](https://support.apple.com/en-ca/guide/security/secb9cb3140c/web), I stumbled upon the followng:

```
In iOS 14 or later, iPadOS 14 or later, and watchOS 7 or later, when an iPhone, iPad, iPod touch, or Apple Watch connects to a Wi-Fi network, it identifies itself with a unique (random) MAC address per network. This feature can be disabled either by the user or using a new option in the Wi-Fi payload. Under certain circumstances, the device will fall back to the actual MAC address.
```

The important part being `Under certain circumstances, the device will fall back to the actual MAC address`. If we can identify the necessary conditions, we may be able to force these devices just to use their real MAC addresses. MAC randomization is likely disabled when an access point uses MAC screening (i.e only particular devices are allowed to use the network). If this is the case, the device must know that the AP uses MAC screening. The first option is that the user must configure it themselves. If this is the case, there is nothing we can do. If, however, the AP somehow transmits that it uses MAC screening in beacons that it periodically sends out, then we could trick devices into revealing their true MAC address using a rogue AP. This is just an aside, and won't be explored further for now.
